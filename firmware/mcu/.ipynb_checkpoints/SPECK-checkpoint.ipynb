```python
from tqdm.notebook import trange
import numpy as np
import time

ktp = cw.ktp.Basic()
ktp.key_len = 16
ktp.text_len = 8 # –ë–ª–æ–∫ SPECK-64

N = 100  # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç—Ä–∞—Å—Å
key, text = ktp.next()

trace_array = []
textin_array = []

for i in trange(N, desc='Capturing traces'):
    key, text = ktp.next() # –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º —Å–ª—É—á–∞–π–Ω—ã–π –∫–ª—é—á –∏ —Ç–µ–∫—Å—Ç
    
    target.set_key(key) # –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –∫–ª—é—á (–≤ simpleserial-crypto)
    
    scope.arm()
    
    target.simpleserial_write('p', text)
    
    ret = scope.capture()
    if ret:
        print("Target timed out!")
        continue
        
    response = target.simpleserial_read('r', ktp.text_len)
    
    trace_array.append(scope.get_last_trace())
    textin_array.append(text)

trace_array = np.array(trace_array)
textin_array = np.array(textin_array)
```

```python
%matplotlib notebook
import matplotlib.pylab as plt

plt.figure()
plt.plot(trace_array[0], 'r')
plt.plot(trace_array[1], 'g')
plt.show()
```

```python
# –¢–∞–±–ª–∏—Ü–∞ –≤–µ—Å–æ–≤ –•—ç–º–º–∏–Ω–≥–∞
HW = [bin(n).count("1") for n in range(0, 256)]

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è SPECK-64/128 (32-–±–∏—Ç–Ω—ã–µ —Å–ª–æ–≤–∞)
alpha = 8
beta = 3
MASK = 0xFFFFFFFF

def rotr(x, n):
    return ((x >> n) | (x << (32 - n))) & MASK

# –≠—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è –≤—ã—á–∏—Å–ª—è–µ—Ç –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –º—ã –±—É–¥–µ–º –∞—Ç–∞–∫–æ–≤–∞—Ç—å
def speck_intermediate(pt, k_guess_byte, bytenum):
    """
    –í—ã—á–∏—Å–ª—è–µ—Ç –≥–∏–ø–æ—Ç–µ—Ç–∏—á–µ—Å–∫–∏–π –±–∞–π—Ç –ø–æ—Å–ª–µ XOR —Å –∫–ª—é—á–æ–º –≤ –ø–µ—Ä–≤–æ–º —Ä–∞—É–Ω–¥–µ.
    pt: 8-–±–∞–π—Ç–Ω—ã–π plaintext
    k_guess_byte: 1-–±–∞–π—Ç–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –∫–ª—é—á–∞
    bytenum: –Ω–æ–º–µ—Ä –∞—Ç–∞–∫—É–µ–º–æ–≥–æ –±–∞–π—Ç–∞ –≤ –∫–ª—é—á–µ (0-3 –¥–ª—è k[0])
    """
    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º plaintext –≤ –¥–≤–∞ 32-–±–∏—Ç–Ω—ã—Ö —Å–ª–æ–≤–∞ (little-endian)
    y = pt[0] | (pt[1] << 8) | (pt[2] << 16) | (pt[3] << 24)
    x = pt[4] | (pt[5] << 8) | (pt[6] << 16) | (pt[7] << 24)
    
    # –ü–µ—Ä–≤–∞—è —á–∞—Å—Ç—å –ø–µ—Ä–≤–æ–≥–æ —Ä–∞—É–Ω–¥–∞ (–¥–æ XOR —Å –∫–ª—é—á–æ–º)
    x_rot = rotr(x, alpha)
    x_add = (x_rot + y) & MASK
    
    # –ò–∑–≤–ª–µ–∫–∞–µ–º –±–∞–π—Ç –∏–∑ –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω–æ–≥–æ —Å–æ—Å—Ç–æ—è–Ω–∏—è
    intermediate_byte = (x_add >> (bytenum * 8)) & 0xFF
    
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç XOR
    return intermediate_byte ^ k_guess_byte
```

```python
import numpy as np

t_bar = np.mean(trace_array, axis=0)
o_t = np.std(trace_array, axis=0)

recovered_key = [0] * 4 # –í–æ—Å—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º 4 –±–∞–π—Ç–∞ k[0]
max_corr_values = [0] * 4

for bnum in range(4): # –ê—Ç–∞–∫—É–µ–º –±–∞–π—Ç—ã 0, 1, 2, 3
    max_corr = -1
    best_guess = 0
    
    for kguess in trange(256, desc=f'Bruteforcing byte {bnum}'):
        # –í—ã—á–∏—Å–ª—è–µ–º –≥–∏–ø–æ—Ç–µ—Ç–∏—á–µ—Å–∫–∏–µ –≤–µ—Å–∞ –•—ç–º–º–∏–Ω–≥–∞ –¥–ª—è –≤—Å–µ—Ö —Ç—Ä–∞—Å—Å
        hws = np.array([HW[speck_intermediate(pt, kguess, bnum)] for pt in textin_array])
        
        # –ü—Ä–æ–ø—É—Å–∫–∞–µ–º –≥–∏–ø–æ—Ç–µ–∑—ã, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –≤—ã–∑—ã–≤–∞—é—Ç –∏–∑–º–µ–Ω–µ–Ω–∏–π (std dev = 0)
        if np.std(hws) == 0:
            continue
        
        # –í—ã—á–∏—Å–ª—è–µ–º –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é
        hws_bar = np.mean(hws)
        o_hws = np.std(hws)
        
        correlation = np.sum((trace_array - t_bar) * (hws - hws_bar).reshape(N, 1), axis=0)
        cpa_output = correlation / (o_t * o_hws * N)
        
        # –ù–∞—Ö–æ–¥–∏–º –ø–∏–∫ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
        peak_corr = max(abs(cpa_output))
        
        if peak_corr > max_corr:
            max_corr = peak_corr
            best_guess = kguess
            
    recovered_key[bnum] = best_guess
    max_corr_values[bnum] = max_corr
    
print(f"\n–í–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—ã–µ –±–∞–π—Ç—ã k[0]: {[hex(b) for b in recovered_key]}")
print(f"–ü–∏–∫–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏: {max_corr_values}")
```

```python
# –ò–∑–≤–µ—Å—Ç–Ω—ã–π –∫–ª—é—á –∏–∑ –ø–æ—Å–ª–µ–¥–Ω–µ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ —Å–±–æ—Ä–∞ —Ç—Ä–∞—Å—Å
known_key = ktp.get_last_key() 

# –í simpleserial-crypto –¥–ª—è SPECK-64/128 k[0] - —ç—Ç–æ –ø–µ—Ä–≤—ã–µ 4 –±–∞–π—Ç–∞ –º–∞—Å—Ç–µ—Ä-–∫–ª—é—á–∞
expected_key_part = known_key[0:4]
recovered_key_bytes = bytearray(recovered_key)

print(f"–û–∂–∏–¥–∞–µ–º—ã–µ –±–∞–π—Ç—ã: {expected_key_part.hex()}")
print(f"–í–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—ã–µ –±–∞–π—Ç—ã: {recovered_key_bytes.hex()}")

if expected_key_part == recovered_key_bytes:
    print("‚úî –£—Å–ø–µ—Ö! –ö–ª—é—á –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω –≤–µ—Ä–Ω–æ.")
else:
    print("‚ùå –û—à–∏–±–∫–∞: –ö–ª—é—á –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω –Ω–µ–≤–µ—Ä–Ω–æ.")

# –û—á–∏—Å—Ç–∫–∞
scope.dis()
target.dis()
```
